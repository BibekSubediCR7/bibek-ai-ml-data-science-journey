# Online Retail Messy Dataset
This dataset is collected from kaggle. It is easy level of data cleaning where I'll apply my pandas skills to clean the data. It is real world messy dataset. Lets get started.

### Let's import pandas and numpy

```python
import numpy as numpy
import pandas as pd
```

Now lets load the dataset using pandas

```python
df=pd.read_csv("online_retail_real_world.csv")
df
```

Lets see the data now. What each column has , rows, stats summary and everything about the data.

```python
df.sample(10)
```

```python
display(df.dtypes) #checking data types of cols
display(df.info()) #general info about data.
df.describe() #stats summary of numeric value
```

```python
#making a copy of the file first so the original is not modified.
dfc=df.copy()
dfc.head()
```

```python
#Renaming columns to match the snakecase pattern
dfc=dfc.rename(columns={"OrderID":"order_id",
                       "CustomerID":"customer_id",
                       "ProductName":"product_name",
                       "Brand":"brand",
                       "Raw_Weight":"raw_weight",
                       "Country":"country",
                       "OrderDate":"order_date",
                       "UnitPrice":"unit_price"})
dfc.columns.tolist()

# We could have used df=df.columns("col1",col2,..) as well with changed col names or used regex.
```

```python
dfc.sample(10)
```

```python
#Finding Null values sum of each col
print(dfc.isna().sum())
(dfc.isna().sum()/len(df))*100 #percentage of na
```

```python
#Deleting the empty rows which has product name. 
dfc.dropna(subset=['product_name'],inplace=True)
dfc.isna().sum()
```

```python
#Filling na values with mean of unit_price, cause dats is normally distributed.
#dfc["unit_price"]=dfc["unit_price"].fillna(dfc["unit_price"].mean())
dfc.isna().sum()
dfc["unit_price"].skew() #Negatively skewed only a bit.
```

```python
# Lets see one more random rows to find out what to fix in our data.
dfc.sample(20)
```

So the average_price has defects in its value. It's messy, differerent units, and extra spaces and many things. 
I"m gonna use regex to replace all which are not numbers and decimals with "".

```python
#Cleaning the avg_price column mess with regex
dfc['raw_weight'] = dfc['raw_weight'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
dfc.sample(20)
```

```python
dfc.isnull().sum()
dfc.describe()
```

```python
#checking the values of raw_weight
display(dfc[dfc["raw_weight"]] <=1)
```

```python
#Deleting the rows which has less than 1 value cause it doesnt make any sense.
# Keep only rows where raw_weight is 1 or more
dfc = dfc[dfc['raw_weight'] >= 1]
# we could have used dfc.drop(dfc[dfc'raw_weight'<=1],inplace=True).. this works as well.
```

```python
print(dfc.info())
dfc.describe()
```

```python
dfc.sample(20)
```

I see that brand name and product name has brand name, manufacturer, company like that. I'll only keep the first one.

```python
#I'm only keeping first part of brand.
#splits the product into a list and takes the first element and strips it.
dfc['product_name']=dfc["product_name"].str.split(",").str[0].str.strip()

dfc['brand']=dfc["brand"].str.split(",").str[0].str.strip()
dfc.sample(10)
```

The product name has some numbers * as well lets only keep the text in there.

```python
dfc["product_name"]=dfc["product_name"].replace(r'[^a-zA-Z\s]','',regex=True).str.strip()
dfc
```

```python
dfc.sample(10)
dfc.isnull().sum()
```

```python
dfc["brand"].nunique()
dfc.shape
(dfc.isnull().sum()/len(df))*100
(dfc[dfc["brand"].isnull()])
```

```python
#Lets  fill the null brand names with "Unknown"
dfc["brand"]=dfc["brand"].fillna("Unknown")
```

```python
dfc.dtypes
```

Lets convert float cols to int

```python
dfc[["raw_weight","unit_price"]]=dfc[["raw_weight","unit_price"]].astype(int)
```

```python
dfc.dtypes
```

```python
#checking if there are duplicates order id
dfc.duplicated(subset="order_id").sum()
```

```python
#checking if there are duplicated customer id
dfc.duplicated(subset="customer_id").sum()
```

```python
#checking both order and customer id.
dfc.duplicated(subset=["order_id","customer_id"]).sum()
```

```python
#lets check the order_id duplication.other dups are normal.
dfc[dfc.duplicated(subset="order_id",keep=False)].sort_values("order_id")
```

```python
#Deleting dubs order id all because it doesn't match with other column, so its an error from the system generating order ids.
dfc.drop_duplicates(subset="order_id")
```

```python
dfc.info()
```

### Handling errors in country names.
### I'll use the first country name, remove en: at the beginning

```python
dfc.sample(10)
```

```python
dfc['country'] = dfc['country'].str.split(',').str[0] # taking the first country.
dfc["country"]=dfc["country"].str.replace("en:","",case=False)
```

```python
dfc.sample(20)
```

```python
#Lowercasing brand, country, product_name rows
dfc["product_name"]=dfc["product_name"].str.lower()
dfc["country"]=dfc["country"].str.lower()
dfc["brand"]=dfc["brand"].str.lower()
```

```python
#reseting index.
dfc.reset_index(inplace=True,drop=True)
dfc.drop(columns="index")
```

```python
#inspecting the data after cleaning.

display(dfc.head(20))
display(dfc.info())
display(dfc.shape)
display(dfc.describe())
display(dfc.isnull().sum())
```

```python
dfc.drop(columns="index",inplace=True)
dfc
```
